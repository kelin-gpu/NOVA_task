```python
prompt = f"""
角色设定：
你是一位信息整理助手，擅长把零散的通知整合成按时间顺序排列的时间轴摘要。你会保留时间、事件和简短说明，使阅读者能快速了解每个阶段发生了什么。

任务说明：
我将提供一份包含多个时间段通知的合并文本，这些内容顺序混乱、部分重复或跨年份。
请你将它整理成一份**“时间线摘要”**，要求如下：

按时间顺序排列（从最早到最新）。

每条保留时间与事件主题（如“2022年8月 – 开学返校安排发布”）。

每条用一句或几句简短描述说明内容重点。

不合并不同时间的事件，即使主题相似（例如两次体测通知，也要分别写出）。

保持生活化语气，像发在群里的时间线总结贴，不要太正式。

格式整齐，推荐如下结构：

## 🗓️ 2022年8月
- **8月13日–8月18日**：发布开学返校安排，老生9月5日开课……
- **8月19日**：更新VPN、核酸检测、图书馆开放信息……

## 🗓️ 2022年9月
- **9月14日–9月20日**：补退选开放、体育课名额更新……
- **9月25日**：学费扣款、通行权限调整……

## 🗓️ 2023年2月–3月
- 校园网不稳定问题出现；食堂与校医院恢复营业……

## 🗓️ 2023年9月
- 新学期开学通知、热水与电费说明、打印机更新……

（以此类推）
"""
```

```python
def call_qwen_model(message):
    # 初始化客户端
    client = OpenAI(
        api_key=os.getenv("DASHSCOPE_API_KEY"),
        base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
    )

    # 调用模型
    completion = client.chat.completions.create(
        model="qwen-plus",  # 也可以尝试 "qwen-turbo" 或 "qwen-max"
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content":  f"以下是聊天记录，请生成通知摘要：\n\n{message}"},
        ],
        temperature=0.5,
        max_tokens=32768
    )

    # 提取模型输出
    x = completion.choices[0].message.content.strip()
    return x
```

```python
if __name__ == "__main__":
    # === 1️⃣ 从 txt 文件中读取内容 ===
    txt_path = fr"D:\google_dl\Nova\nova_qq_message\qq_message_txt_to_other\combine.txt"  # 你的 txt 文件路径
    with open(txt_path, "r", encoding="utf-8") as f:
        message = f.read().strip()
    # === 2️⃣ 调用模型 ===
    result = call_qwen_model(message)
```
